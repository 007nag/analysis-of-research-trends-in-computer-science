```python
import json
import matplotlib.pyplot as plt
```

```python
f = open('dblpv13.json','r', encoding='utf-8')
year_dict = {}
count = 0
while True:
    count += 1
 

    line = f.readline()
 

    if not line:
        break

    if(line.find('"year" :') != -1):
        year = line.strip()
        year = year[19:-2]
        if int(year)<1800 or int(year)>2022:
            print(line)
        if year not in year_dict.keys():
            year_dict[year] = 1
        else :
            year_dict[year] +=1
f.close()
```

```python
year_dict=dict(sorted(year_dict.items()))
```

```python
year_dict
```

```python
delete_list = []
for i in year_dict.keys():
    if(len(i) == 0):
        delete_list.append('')
        continue
    else:
        i = int(i)
        if i<1800 or i>2022:
            delete_list.append(str(i))
```

```python
for i in delete_list:
    del year_dict[i]
```

```python
yt=[]
x= 1800
while x<2030:
    yt.append(x)
    x = x + 10
```

```python
plt.rcParams["figure.figsize"] = (30,20)
plt.rcParams.update({'font.size': 20})
```

```python
testx_list = [int(i) for i in list(year_dict.keys())]
plt.plot(testx_list,list(year_dict.values()))

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('year wise distribution of papers')
plt.xticks(yt)

plt.show()
```
